```python
# Merge
df1 = pd.DataFrame({'lkey': ['foo', 'bar', 'baz', 'foo'],
                    'value': [1, 2, 3, 5]})
df2=pd.DataFrame({'rkey': ['foo', 'bar', 'baz', 'foo'],
                    'value': [5, 6, 7, 8]})

print(df1)
print("\n")
print(df2)
print("\n")

pd.merge(df1,df2,left_on='lkey',right_on='rkey',suffixes=('_left','_right'))
```

```python
df1 = pd.DataFrame({'a': ['foo', 'bar'], 'b': [1, 2]})
df2 = pd.DataFrame({'a': ['foo', 'baz'], 'c': [3, 4]})

print(df1)
print("\n")
print(df2)
print("\n")

print(pd.merge(df1,df2, on='a'))
print("\n")

print(pd.merge(df1,df2,how='right',on='a'))
print('\n')

print(pd.merge(df1,df2,how='outer',on='a'))

pd.merge(df1,df2,left_on='a',right_on='a',suffixes=('_left','_right'))
```

```python
# How do you perform an inner join using merge?
import pandas as pd
df1=pd.DataFrame({"key":["a","b","c"],"value":[1,2,3]})
print(df1)

df2=pd.DataFrame({"key":["a","b","c"],"value":[4,5,6]})
print(df2)

pd.merge(df1,df2,how="inner",on="key")

```

```python
# how to perform outer join
import pandas as pd
df1=pd.DataFrame({"key":["a","b","c"],"value":[1,2,3]})
print(df1)

df2=pd.DataFrame({"key":["b","c","d"],"value":[4,5,6]})
print(df2)

pd.merge(df1,df2,how="outer",on="key")
```

```python
import pandas as pd

# Create two DataFrames with overlapping column names
df1 = pd.DataFrame({
    'key': ['A', 'B', 'C', 'D'],
    'value': [1, 2, 3, 4],
    'data': ['X1', 'X2', 'X3', 'X4']
})

df2 = pd.DataFrame({
    'key': ['B', 'D', 'E', 'F'],
    'value': [5, 6, 7, 8],
    'data': ['Y1', 'Y2', 'Y3', 'Y4']
})

# Merge the DataFrames using the 'key' column
merged_df = pd.merge(df1, df2, on='key', suffixes=('_left', '_right'))

print(merged_df)

print("\n")

result = pd.merge(df1, df2, on='key', how='outer', indicator=True)
print(result)
```

Question 1: Merge Two DataFrames
Problem Statement: Given two DataFrames df1 and df2, merge them on a common column id. Perform an inner join and return the merged DataFrame.

```python
import pandas as pd

# Example DataFrames
df1 = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'value1': ['A', 'B', 'C', 'D']
})

df2 = pd.DataFrame({
    'id': [3, 4, 5, 6],
    'value2': ['X', 'Y', 'Z', 'W']
})


result= pd.merge(df1,df2,on="id",how="inner")
print(result)
```

Question 3: Merge with Suffixes
Problem Statement: Given two DataFrames df1 and df2 with overlapping column names, merge them on a common column key and use suffixes _left and _right to distinguish columns from df1 and df2. Perform a left join and return the merged DataFrame.

```python
import pandas as pd

# Example DataFrames
df1 = pd.DataFrame({
    'key': ['K0', 'K1', 'K2'],
    'value': [1, 2, 3]
})

df2 = pd.DataFrame({
    'key': ['K1', 'K2', 'K3'],
    'value': [4, 5, 6]
})


pd.merge(df1, df2, on='key',suffixes=('_left','_right'),how="left")
```

Question 4: Concatenate DataFrames with Keys
Problem Statement: Given two DataFrames df1 and df2, concatenate them vertically and add keys to identify each DataFrame. Return the concatenated DataFrame with a hierarchical index.

```python
import pandas as pd

# Example DataFrames
df1 = pd.DataFrame({
    'A': ['A0', 'A1'],
    'B': ['B0', 'B1']
})

df2 = pd.DataFrame({
    'A': ['A2', 'A3'],
    'B': ['B2', 'B3']
})
result=pd.concat([df1,df2],keys=["df1","df2"])
result
```

Question 5: Merge on Index
Problem Statement: Given two DataFrames df1 and df2, merge them on their indices. Perform an outer join and return the merged DataFrame.

```python
import pandas as pd

# Example DataFrames
df1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2']
}, index=['K0', 'K1', 'K2'])

df2 = pd.DataFrame({
    'B': ['B0', 'B1', 'B2']
}, index=['K0', 'K2', 'K3'])


print(df1)
print('\n')
print(df2)

result=pd.merge(df1,df2,left_index=True,right_index=True,how="inner")
result
```
